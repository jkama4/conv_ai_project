```python
def start_conversation(
    history: list[dict],
    n: int = 5
) -> list[dict]:
        
    for i in range(n):
        assistant_response = assistant_agent._call(history=history)
        history.append({
            "role": "assistant",
            "content": assistant_response,
        })

        user_response = traveler_agent._call(history=history)
        history.append({
            "role": "user",
            "content": user_response,
        })

        if "<END>" in history[-1]["content"]:
            return history
        
    return history

# convo = start_conversation(history=history)
```
