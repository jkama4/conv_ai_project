```python
def start_conversation(
    history: list[dict],
    n: int = 5
) -> list[dict]:
        
    for i in range(n):
        assistant_response = assistant_agent._call(history=history)
        history.append({
            "role": "assistant",
            "content": assistant_response,
        })

        user_response = traveler_agent._call(history=history)
        history.append({
            "role": "user",
            "content": user_response,
        })

        if "<END>" in history[-1]["content"]:
            return history
        
    return history

# convo = start_conversation(history=history)
```

```python
from pathlib import Path
import json
import torch

from typing import List, Dict
```

```python
def tensor_to_python(obj):
    if isinstance(obj, torch.Tensor):
        return obj.item()
    return obj
```

```python
def save_evaluations(
    evaluated_conversations: List[Dict],
    custom: bool = False
) -> None:


    save_file: Path = Path("/Users/jaydenkm/workspaces/school/conversational_ai/final_project/src/final_project/data/evaluated_conversations/custom_evaluated.json")

    with open(save_file, "w") as f:
        json.dump(evaluated_conversations, f, default=tensor_to_python)
    
    return None
```

```python
save_evaluations(
    evaluated_conversations=evaluated_conversations,
    custom=True,
)
```
